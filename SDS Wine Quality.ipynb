# White Wine Quality Dataset

## Imports

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH = "winequality-white.csv"
DATA = pd.read_csv(PATH,sep=";")
```

### Data Types

```python
DATA.describe()
```

### Histogram

```python
for ax in DATA.hist(figsize=(20,20)):
    ax
```

```python
plt.figure(figsize=(20,100))
for ind,name in enumerate(DATA.columns):
    plt.subplot(len(DATA.columns),2,ind+1)
    plt.title(name)
    plt.boxplot(DATA[name])
plt.show()
```
